The above plots doesnot make sense because it is not possible for count more than 250+ count to have 0% of the total electricity access. Therefore changing the zeros to the mean of the respective columns

```python
# Function to calculate the mean of non-zero values for each column
def mean_non_zero(series):
    non_zero_values = series[series != 0]
    if len(non_zero_values) == 0:
        return 0
    else:
        return non_zero_values.mean()

numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns

# Calculate the mean of non-zero values for each column
means = data[numeric_cols].apply(mean_non_zero)

# Replace zero values with the calculated mean for each column
data[numeric_cols] = data[numeric_cols].apply(lambda x: x.replace(0, means[x.name]))
```

1. How has the overall electricity access evolved over time in Sub-Saharan Africa compared to other regions?

* Plot Suggestion: Line chart comparing el_access_total over the years for Sub-Saharan Africa vs. other regions.

* Analysis: This will help identify trends and patterns in electricity access across different regions and over time.


```python
import matplotlib.pyplot as plt
import seaborn as sns

def perform_eda(data, palette='viridis'):
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=data, x='year', y='el_access_total', hue='entity', palette=palette, legend='full')
    plt.title('Total Electricity Access Over the Years for All Countries')
    plt.xlabel('Year')
    plt.ylabel('Total Electricity Access (%)')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Assuming 'data' contains the entire dataset and 'unique_entities' is a list of all unique entities
perform_eda(data)
```

5. What is the composition of electricity generation sources in Sub-Saharan Africa compared to other regions?

```python
supply_columns = ['el_from_gas', 'el_from_nuclear', 'el_from_hydro',
                  'el_from_solar', 'el_from_oil', 'el_from_wind', 'el_from_bioenergy',
                  'renewables_other']

demand_column = 'el_access_total'
```

```python
# Calculate total supply for each region
total_supply = data.groupby('entity')[supply_columns].sum()

# Calculate total demand for each region
total_demand = data.groupby('entity')[demand_column].sum()

total_supply.reset_index(inplace=True)

total_demand = total_demand.reset_index()

# Display the total supply by region
print("Total Supply by Region:")
print(total_supply)

print("Total Demand by Region:")
print(total_demand)
```

```python
total_supply = data.groupby('entity')[supply_columns].sum()
percentages = total_supply.div(total_supply.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
percentages.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Percentage Breakdown of Electricity Supply Sources by Region')
plt.xlabel('Region')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=90)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
# Calculate renewable and non-renewable energy for all regions
data['renewable_energy'] = data[['el_from_solar', 'el_from_wind', 'el_from_bioenergy', 'renewables_other']].sum(axis=1)
data['non_renewable_energy'] = data[['el_from_coal', 'el_from_gas', 'el_from_oil', 'el_from_nuclear']].sum(axis=1)
```

```python
renewable_vs_non_renewable_all_regions = data.groupby('entity')[['renewable_energy', 'non_renewable_energy']].sum()

# Bar plot for renewable vs non-renewable energy supply for all regions
renewable_vs_non_renewable_all_regions.plot(kind='bar', figsize=(14, 8))
plt.title('Renewable vs Non-Renewable Energy Supply for All Regions')
plt.xlabel('Region')
plt.ylabel('Total Energy Supply')
plt.xticks(rotation=90)
plt.legend(title='Energy Type')
plt.show()
```
