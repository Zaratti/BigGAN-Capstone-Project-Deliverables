2. What is the distribution of electricity access between urban and rural areas in Sub-Saharan Africa?

```python
urban_access = data.groupby('year')['el_access_urban'].mean().reset_index()
rural_access = data.groupby('year')['el_access_rural'].mean().reset_index()

# Plot the data
plt.figure(figsize=(8, 4))
plt.plot(urban_access['year'], urban_access['el_access_urban'], label='Urban', marker='o')
plt.plot(rural_access['year'], rural_access['el_access_rural'], label='Rural', marker='o')
plt.xlabel('Year')
plt.ylabel('Electricity Access (%)')
plt.title('Electricity Access in Urban vs Rural Areas in Sub-Saharan Africa')
plt.legend()
plt.grid(True)
plt.show()
```

The difference in the Electricity access for urban and rural is clear. The urban stands at more than 50% and increasing with time whereas the rural is at a little over 20% with slight increase with time

3. What is the relationship between GDP per capita and electricity access in different regions?

```python
# Scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(data['gdp_pc'], data['el_access_total'], marker='o', color='blue', alpha=0.5)
plt.xlabel('GDP per Capita')
plt.ylabel('Electricity Access (%)')
plt.title('Relationship between GDP per Capita and Electricity Access')
plt.grid(True)
plt.show()
```

4. How does the population without electricity access vary across regions over time?

```python
plt.figure(figsize=(10, 6))

grouped_data = data.groupby(['entity', 'year'])['pop_no_el_access_total'].sum().reset_index()

for region, region_data in grouped_data.groupby('entity'):
    plt.plot(region_data['year'], region_data['pop_no_el_access_total'], marker='o', label=region)

plt.xlabel('Year')
plt.ylabel('Population without Electricity Access')
plt.title('Population without Electricity Access Across Regions Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Group by 'entity' (country) and calculate the sum of 'pop_no_el_access_total'
grouped_data = data.groupby('entity')['pop_no_el_access_total'].sum().reset_index()

# Sort the grouped data in descending order based on the summed values
sorted_data = grouped_data.sort_values(by='pop_no_el_access_total', ascending=False)

# Select the top 5 regions with the maximum summed values
top_regions = sorted_data.head(5)

top_region_names = top_regions['entity'].tolist()
print("Top 5 regions with the maximum population without electricity access:")
for region in top_region_names:
    print(region)
```

6. How does the installed electricity capacity vary between Sub-Saharan Africa countries?

```python
plt.figure(figsize=(10, 6))

bar_width = 0.4
bars = plt.bar(data['entity'], data['el_inst_capacity_total'], color='blue', width=bar_width, label='Sub-Saharan Africa')

plt.xlabel('Regions')
plt.ylabel('Installed Electricity Capacity')
plt.title('Comparison of Installed Electricity Capacity between Sub-Saharan Africa and Other Regions')
plt.legend()
plt.xticks(rotation=90)
plt.ylim(0, 35000)  # Setting y-axis limit for better scaling
plt.tight_layout()
plt.show()
```

```python
sum_capacity = data.groupby('entity')['el_inst_capacity_total'].sum()
top_5_high = sum_capacity.nlargest(5)
top_5_low = sum_capacity.nsmallest(5)

print("Top 5 Countries with Highest Installed Electricity Capacity:")
print(top_5_high)
print("\nTop 5 Countries with Lowest Installed Electricity Capacity:")
print(top_5_low)
```

Inference: top 5 countries which has the highest installed Electricity capacity over the years and the lowest as well.
